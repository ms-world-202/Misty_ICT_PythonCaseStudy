```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('/content/iris (1).xls')
df.head()
```

#Fill missing columns with mean value

```python
df.fillna({'SL': df['SL'].mean()},inplace=True)
df.fillna({'SW': df['SW'].mean()},inplace=True)
df.fillna({'PL': df['PL'].mean()},inplace=True)
```

```python
df.isnull().sum()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,5), dpi = 100)
fig.suptitle('Distribution of (sepal length, sepal width, petal length, petal width) per Species')

# Distribution of sepal length per Species
sns.kdeplot(ax = axes[0,0], data = df, x = 'SL', hue = "Classification", alpha = 0.5, shade = True)
axes[0,0].set_xlabel("Sepal Length CM")
axes[0,0].get_legend().remove()

# Distribution of sepal width per Species
sns.kdeplot(ax = axes[0,1], data = df, x = 'SW', hue = "Classification", alpha = 0.5, shade = True)
axes[0,1].set_xlabel("Sepal width CM")
axes[0,1].get_legend().remove()

# Distribution of petal length per Species
sns.kdeplot(ax = axes[1,0], data = df, x = 'PL', hue = "Classification", alpha = 0.5, shade = True)
axes[1,0].set_xlabel("Petal Length CM")
axes[1,0].get_legend().remove()

# Distribution of petal width per Species
sns.kdeplot(ax = axes[1,1], data = df, x = 'PW', hue = "Classification", alpha = 0.5, shade = True)
axes[1,1].set_xlabel("Petal Width CM")

plt.tight_layout()
```

```python
# Scatter plot of PL vs PW
plt.figure(figsize = (7, 3), dpi = 100)
sns.scatterplot(data = df, x = 'PL', y = 'PW', hue = "Classification")
plt.title("Classification clusters based on Sepal length and width")
plt.xlabel("Petal Length Cm")
plt.ylabel("Petal Width Cm")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

From the above scatter plot, we understand that iris-setosa can be clearly distinguished by petal length and petal width and same goes for the other two as well.

```python
# Scatter plot of SL vs SW
plt.figure(figsize = (7, 3), dpi = 100)
sns.scatterplot(data = df, x = 'SL', y = 'SW', hue = "Classification")
plt.title("Classification clusters based on Sepal length and width")
plt.xlabel("Sepal Length Cm")
plt.ylabel("Sepal Width Cm")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Sepal length and sepal width is not as good as petal length and petal width to do the classification.

```python
#box plots
fig, axes = plt.subplots(2, 2, figsize=(10,5), dpi = 100)

#Mean Sepal Length
sns.boxplot(ax = axes[0,0], data = df, x = "Classification", y = 'SL',hue="Classification")
axes[0,0].set_xlabel(None)
axes[0,0].set_ylabel(None)
axes[0,0].set_title("Mean Sepal Length")

#Mean Sepal Width
sns.boxplot(ax = axes[0,1], data = df, x = "Classification", y = 'SW',hue="Classification")
axes[0,1].set_xlabel(None)
axes[0,1].set_ylabel(None)
axes[0,1].set_title("Mean Sepal Width")

#Mean Petal Length
sns.boxplot(ax = axes[1,0], data = df, x = "Classification", y = 'PL',hue="Classification")
axes[1,0].set_xlabel(None)
axes[1,0].set_ylabel(None)
axes[1,0].set_title("Mean Petal Length")

#Mean Petal Width
sns.boxplot(ax = axes[1,1], data = df, x = "Classification", y = 'PW',hue="Classification")
axes[1,1].set_xlabel(None)
axes[1,1].set_ylabel(None)
axes[1,1].set_title("Mean Petal Width")

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)
```

```python
x = df.drop("Classification", axis = 1)
y = df["Classification"]
```

#Splitting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Scaled_x_train = scaler.fit_transform(x_train)
Scaled_x_test = scaler.transform(x_test) #transform the test data
```

# Logistic Regression

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(Scaled_x_train, y_train)
```

```python
#predictions
y_pred = log_model.predict(Scaled_x_test)
```

```python
# metrics
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

#Naive Bayes classifiers

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(Scaled_x_train, y_train)

#predict
y_pred_gnb = gnb.predict(Scaled_x_test)

#metrics
print(confusion_matrix(y_test,y_pred_gnb))
print(classification_report(y_test,y_pred_gnb))
print(accuracy_score(y_test,y_pred_gnb))
```
